```python
import seaborn as sns 
tips=sns.load_dataset("tips")
tips
```

```python
fig =  go.Figure()
fig.add_trace(go.Scatter(x=tips.total_bill,y=tips.tip, mode='markers'))
```

```python
fig.add_trace(go.Scatter(x=tips.total_bill,y=tips.tip, mode='lines'))
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(x=[1,2,3,4,5], y=[3,8,5,2,6]))
```

```python
fig = go.Figure()
```

```python
fig = go.Figure()
fig.add_trace(go.Histogram(x=tips['total_bill']))
```

```python
fig = go.Figure()
#the bubble is directly proportional to marker size 
fig.add_trace(go.Scatter(x=tips.total_bill, y= tips.tip, mode = 'markers', marker_size = 12* tips['size']))
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter3d(x=tips.total_bill, y= tips.tip, mode='markers', marker_size = 10*tips['size'], z = tips['size'])) 
```

```python
#using plotly express
x=[1,2,3,4]
y=[10,13,15,18]

fig= px.scatter(x,y)
fig.show()
```

```python
fig= px.line(x,y)
fig.show()
```

```python
x=['A','B','C']
y= [5,10,3]
fig= px.bar(x=x,y=y)
fig.show()
```

```python
x=[5,10,3,4,5,6,6,10,10,11,11,12]
fig= px.histogram(x)
fig.show()
```

```python
x=['A','B','C']
y=[5,10,3,25,12,33,100]
fig = px.box(x,y)
fig.show()
```

```python
data = [[1,2,3],[4,5,6],[7,8,9]]
fig = px.imshow(data)
fig.show()
```

```python
#bubble chart
x=[1,2,3,4]
y=[5,8,12,3]
fig = px.scatter(x,y,size=[30,10,50,200])
```

```python
#3d chart 
x=[1,2,3,4]
y=[5,8,12,3]
z=[3,4,5,6]
fig = px.scatter_3d(x=x,y=y,z=z)
fig.show()
```
