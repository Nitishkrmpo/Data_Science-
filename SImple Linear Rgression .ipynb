```python
data=pd.DataFrame(diabetes.data,columns=diabetes.feature_names)
```

```python
data['target']=diabetes.target
```

```python
##EDA DATA CLEANING ,DATA PREPARATION ,FEATURE ENGINEERING   ALRADY DONE 
```

```python
#DIVIDE THE DATA IN X & Y
```

```python
#simiple Linear REgression using one  dataset namely as bmi and target 
```

```python
x=data[["bmi"]]
y=data["target"]
```

```python
x
```

```python
y
```

```python
#split the data form tarin Test Split 
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2 , random_state=1)
```

```python
from sklearn.linear_model import LinearRegression 
```

```python
model = LinearRegression
```

```python
#instaniate as LR Model
```

```python
model=LinearRegression()
model
#fit is used to calculate mi and sigma in scalling and tansform is used to apply all these to dataset 
#fit takes  training of the model in Regression Model
#first it will make ordinary least square , take mean square average then it become the mean square error\
#then it is futher optimised using the graident decent alogthim to find out the best theta not and theta 1

#train s used to calutate usingthe optimal  theta not and theta 1  
```

```python
model.fit(x_train, y_train)#at backend , it will use the Gradient decent alogothim to find the optimal solution 
#X_tarin jo hain usko Datframe me pass karna hai in the linear regression model, this is done by passing the double square braket in  during the splitiing and this should not be done for the y Variable  
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred =model.predict(x_test)
```

```python
y_pred
```

```python
#visualize the result 
plt.scatter(x_test, y_test , color = 'black', label ='Actual data')
plt.plot(x_test,y_pred,color='blue',linewidth=3, label="Linear regression model")
plt.xlabel("BMI")
plt.ylabel("One Year Progession_target")
plt.title("Linear Regression on diabetes Data")
plt.legend()
plt.show()
```
